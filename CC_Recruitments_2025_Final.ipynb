#Important Insight 1
This graph just shows that Regular people with rare purchase frequency buy most.
On the other hand Gold Members are least frequent in buying.
While Silver plan members seem to have a more balanced purchase pattern.

```python
#df.groupby('loyalty_status')['purchase_frequency'].value_counts().plot()
import matplotlib.pyplot as plt
df.groupby('loyalty_status')['purchase_frequency'].value_counts().plot()
plt.xticks(rotation=45)  # Rotates x-axis labels for better readability
plt.show()
```
